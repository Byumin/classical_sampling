```python
import pandas as pd
import statsmodels
import pingouin
```

```python
file_path='clean_parent.csv'
sheet_name='clean_parent'
```

```python
data = pd.read_csv(file_path)
```

```python
data
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
```

```python
data['sex'] = data['sex'].astype('category')
data['age'] = data['age'].astype('category')
```

```python
data
```

```python
data.columns
```

```python
# 1) OLS 모형 적합 (연령, 성별, 상호작용 포함)
model = smf.ols('adhd ~ C(age) * C(sex)', data=data).fit()

# 2) Type II ANOVA 실행
aov = anova_lm(model, typ=2)
print(aov)
```

```python
# 3) 오차 제곱합 & 자유도
SS_error = aov.loc['Residual','sum_sq']
df_error = aov.loc['Residual','df']
MS_error = SS_error/df_error

results = {}
for eff in ['C(age)', 'C(sex)', 'C(age):C(sex)']:
    SS_eff = aov.loc[eff, 'sum_sq']
    df_eff = aov.loc[eff, 'df']
    
    # 부분 η²
    eta_p2 = SS_eff / (SS_eff + SS_error)
    
    # 오메가²
    omega2 = (SS_eff - df_eff*MS_error) / (SS_eff + SS_error + MS_error)
    
    # Cohen's f (G*Power 입력용)
    f = np.sqrt(eta_p2/(1-eta_p2))
    
    results[eff] = {"eta_p2": round(eta_p2,3),
                    "omega2": round(omega2,3),
                    "Cohen_f": round(f,3)}

print(results)
```
