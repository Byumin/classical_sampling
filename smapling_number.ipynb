```python
1. 왜 필요한가?

IRB 입장에서는 연구가 과학적·윤리적으로 타당한지 확인해야 합니다.

연구대상자가 너무 적으면 결과의 신뢰성이 떨어지고, 너무 많으면 불필요하게 많은 피험자를 연구에 동원하는 윤리적 문제가 발생합니다.

따라서 “왜 그 수만큼 필요한가?”를 통계적·선행연구적 근거로 제시해야 합니다.

2. 세부 요구사항

문구에서 강조하는 건 크게 3가지입니다.

① 직접 모집하는 경우 → 중도 탈락률 고려

예를 들어, 연구에서 100명이 꼭 필요하다면, 보통 10~20%의 탈락률(중도 포기, 결측 등)을 고려해 120명 정도를 모집해야 합니다.

이걸 명시해야 해요.

예: "최소 표본 수는 100명으로 산출되었으나, 선행연구에서 약 15%의 중도탈락률이 보고되어 총 120명을 모집한다."

② 통계학적 산출 또는 선행연구 근거

표본 수 산출은 G*Power 같은 프로그램을 이용해 **효과크기(effect size), 유의수준(α), 검정력(1-β)**를 설정하여 계산합니다.

또는 “선행연구에서 비슷한 연구는 약 150명을 대상으로 했으므로 동일한 규모를 계획했다”와 같이 씁니다.

③ 필요 최소 표본 수

“가능한 한 많이”가 아니라 **“계획된 연구에서 필요한 최소한”**이어야 합니다.

즉, 통계적으로 유효하면서도 불필요하게 많은 피험자를 모집하지 않는다는 점을 보여줘야 합니다.
```

```python
📌 원리

표본 수 산출은 크게 3가지 변수의 함수로 계산됩니다.

효과 크기 (effect size, d·f²·η² 등)

연구에서 기대하는 차이 또는 관계의 크기.

예: 실험군과 대조군의 평균 차이가 클수록(=효과가 큼) 적은 표본으로도 차이를 발견할 수 있음.

반대로 효과가 작을수록 더 많은 표본이 필요함.

유의수준 (α, 보통 0.05)

“허용하는 1종 오류의 확률(실제로 차이가 없는데 있다고 잘못 결론 내릴 확률)”

α를 낮추면(예: 0.01) 더 엄격해져서 더 많은 표본이 필요.

검정력 (power = 1-β, 보통 0.8 이상)

“실제로 효과가 있을 때, 그것을 올바르게 발견할 확률”

검정력을 높게(예: 0.9) 잡으면, 효과를 놓치지 않으려면 더 많은 표본이 필요.

📌 산정 과정 (예시: G*Power 원리)

연구자가 연구 설계 유형 선택 (예: t-검정, ANOVA, 회귀 등)

선행연구 또는 메타분석을 참고하여 효과 크기 추정 (예: Cohen’s d=0.5, 중간 효과)

유의수준 α (보통 0.05), 검정력 1-β (보통 0.8) 설정

G*Power가 위 값들을 기반으로 “필요 최소 표본 수” 계산

→ 예:

독립표본 t-검정, 효과 크기 d=0.5, α=0.05, power=0.8 → 64명 필요(집단당 32명)

같은 조건에서 효과 크기 d=0.3(작은 효과)로 바꾸면 → 88명 필요(집단당 44명)

즉, 효과가 작거나, 검정을 엄격히 하거나, 높은 신뢰도를 원할수록 표본이 늘어납니다.
```
