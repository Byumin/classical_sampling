```python
file_path='240327_실행기능 규준 재분석(14-15 연령 묶음).xlsx'
sheet_name='표집 데이터'
```

```python
data = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
data.columns = ['age', 'sex', 'tmt1_time', 'tmt1_time(s)', 'tmt1_time(s)_round', 'tmt1_num_error', 'tmt1_color_error', 'tmt1_promotion',
                'tmt2_time', 'tmt2_time(s)', 'tmt2_time(s)_round', 'tmt2_num_error', 'tmt2_color_error', 'tmt2_promotion', 'ratio', 'differ']
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
```

```python
data['age'] = data['age'].astype('category')
```

```python
data
```

```python
# 2) 연령만 포함한 OLS 적합 (one-way ANOVA)
model = smf.ols('tmt1_time ~ C(age)', data=data).fit()
aov = anova_lm(model, typ=2)   # 한 요인이므로 typ=1/2/3 큰 차이 없음
print(aov)

# 3) 제곱합/자유도 추출
SS_between = aov.loc['C(age)', 'sum_sq']
df_between = aov.loc['C(age)', 'df']
SS_within  = aov.loc['Residual', 'sum_sq']
df_within  = aov.loc['Residual', 'df']

# 4) 효과크기 계산
eta2   = SS_between / (SS_between + SS_within)          # 한 요인 ANOVA에서는 partial η² = η²
MS_within = SS_within / df_within
omega2 = (SS_between - df_between * MS_within) / (SS_between + SS_within + MS_within)  # 보수적
f      = np.sqrt(eta2 / (1 - eta2))                     # G*Power용

print({
    "eta2 (연령 주효과)": round(eta2, 3),
    "omega2": round(omega2, 3),
    "Cohen_f": round(f, 3)
})
```

```python
# 2) 연령만 포함한 OLS 적합 (one-way ANOVA)
model = smf.ols('tmt2_time ~ C(age)', data=data).fit()
aov = anova_lm(model, typ=2)   # 한 요인이므로 typ=1/2/3 큰 차이 없음
print(aov)

# 3) 제곱합/자유도 추출
SS_between = aov.loc['C(age)', 'sum_sq']
df_between = aov.loc['C(age)', 'df']
SS_within  = aov.loc['Residual', 'sum_sq']
df_within  = aov.loc['Residual', 'df']

# 4) 효과크기 계산
eta2   = SS_between / (SS_between + SS_within)          # 한 요인 ANOVA에서는 partial η² = η²
MS_within = SS_within / df_within
omega2 = (SS_between - df_between * MS_within) / (SS_between + SS_within + MS_within)  # 보수적
f      = np.sqrt(eta2 / (1 - eta2))                     # G*Power용

print({
    "eta2 (연령 주효과)": round(eta2, 3),
    "omega2": round(omega2, 3),
    "Cohen_f": round(f, 3)
})
```
