```python
import pandas as pd
import statsmodels
import pingouin
```

```python
file_path='작업기억력_표집_원데이터.xlsx'
sheet_name='표집 데이터'
```

```python
data = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm
```

```python
data['만나이'] = data['만나이'].astype('category')
```

```python
data
```

```python
data.columns
```

```python
# 2) 연령만 포함한 OLS 적합 (one-way ANOVA)
model = smf.ols('작업기억지수 ~ C(만나이)', data=data).fit()
aov = anova_lm(model, typ=2)   # 한 요인이므로 typ=1/2/3 큰 차이 없음
print(aov)

# 3) 제곱합/자유도 추출
SS_between = aov.loc['C(만나이)', 'sum_sq']
df_between = aov.loc['C(만나이)', 'df']
SS_within  = aov.loc['Residual', 'sum_sq']
df_within  = aov.loc['Residual', 'df']

# 4) 효과크기 계산
eta2   = SS_between / (SS_between + SS_within)          # 한 요인 ANOVA에서는 partial η² = η²
MS_within = SS_within / df_within
omega2 = (SS_between - df_between * MS_within) / (SS_between + SS_within + MS_within)  # 보수적
f      = np.sqrt(eta2 / (1 - eta2))                     # G*Power용

print({
    "eta2 (연령 주효과)": round(eta2, 3),
    "omega2": round(omega2, 3),
    "Cohen_f": round(f, 3)
})
```
